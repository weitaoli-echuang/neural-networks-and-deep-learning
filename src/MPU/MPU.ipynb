```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import patches
        
```

```python
def weight(d, r):
    d[d>r]=0
    t = 1.5 * d / r
    
    less_mask = (t<0.5) & (t>0)
    less=t[(t<0.5) & (t>0)]
    large_mask=t>0.5
    large=t[t>0.5]
    
    less = -less ** 2 + 0.75
    large=0.5 * (1.5 - large) ** 2

    t[large_mask]=large
    t[less_mask]=less
    return t
```

```python
def f(x,c1,r1,c2,r2):
    d1=np.abs(x-c1)
    d2=np.abs(x-c2)
    w1=weight(d1,r1)
    w2=weight(d2,r2)
    f1=x**2
    f2=(0.5-x)**2
    return w1*f1+w2*f2/(w1+w2)
```

```python
x=np.linspace(0,0.7,20)
y=x**2

x1=np.linspace(0.3,1,20)
y1=(0.5-x1)**2

```

```python
r_1=0.25
r_2=0.25

c_1=0.25
c_2=0.75


x_t=np.linspace(0,1,100)
y_t=f(x_t,c_1,r_1,c_2,r_2)
```

```python
fig = plt.figure(figsize=[20,10])
plt.subplot(211,aspect='equal')
plt.scatter(x,y,color='b')
plt.scatter(x1,y1,color='r')
axis_2=plt.subplot(212,aspect='equal')
axis_2.add_patch(patches.Circle(xy = (c_1, c_1), radius=r_1, alpha=0.5))
axis_2.add_patch(patches.Circle(xy = (c_2, c_1), radius=r_2, alpha=0.5))
axis_2.scatter(x_t,y_t,color='b')
axis_2.scatter(x,y,color='b')
axis_2.scatter(x1,y1,color='r')
plt.show()
```

```python
def f_im(x,y,c1,r1,c2,r2):
    d1=np.sqrt((x-c1[0])**2+(y-c1[1])**2)
    d2=np.sqrt((x-c2[0])**2+(y-c2[1])**2)
    w1=weight(d1,r1)
    w2=weight(d2,r2)
    f1=y-x**2
    f2=y-(0.5-x)**2
    return w1*f1+w2*f2/(w1+w2)
```

```python
c_1=np.array([0.25,0.25])
c_2=np.array([0.75,0.25])
```

```python
x_v=np.linspace(0,1,500)
y_v=np.linspace(0,0.5,500)
x_im,y_im=np.meshgrid(x_v,y_v)
z = f_im(x_im,y_im,c_1,r_1,c_2,r_2)
```

```python
fig = plt.figure(figsize=[20,10])
axis_2=plt.subplot(111,aspect='equal')
axis_2.add_patch(patches.Circle(xy = (c_1[0], c_1[1]), radius=r_1, alpha=0.2))
axis_2.add_patch(patches.Circle(xy = (c_2[0], c_2[1]), radius=r_2, alpha=0.2))
CS=axis_2.contour(x_v,y_v,z,6)
axis_2.clabel(CS,fontsize=10,color='r')
axis_2.scatter(x,y,color='b')
axis_2.scatter(x1,y1,color='r')
axis_2.scatter(x_t,y_t,color='g')
plt.show()
```
